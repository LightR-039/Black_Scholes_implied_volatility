# Black–Scholes Implied Volatility Solver

This notebook demonstrates how to compute **implied volatility** of European options using the Black–Scholes model.

Implied volatility is the volatility value that makes the theoretical Black–Scholes price equal to the observed market price.

Implemented numerical solvers (Newton and Bisection methods) to recover volatility from real option prices.


```python
from src.solvers import implied_volatility
from src.black_scholes import black_scholes_call
from src.visualization import plot_price_vs_volatility
```

```python
import numpy as np
```

## Real-World Scenario

Suppose a trader observes the following market data:

* Stock price: $100
* Strike price: $100
* Time to maturity: 1 year
* Risk-free rate: 2%
* Market option price: $10

 The task is estimate the implied volatility.


```python
S = 100
K = 100
T = 1
r = 0.02
market_price = 10
```

```python
iv, iterations, success = implied_volatility(
    market_price, S, K, T, r
)

print("Implied volatility:", iv)
print("Iterations:", iterations)
print("Converged:", success)
```

```python
model_price = black_scholes_call(S, K, T, r, iv)
print("Model price:", model_price)
```

```python
plot_price_vs_volatility(
    S, K, T, r, market_price
)
```

```python
market_price = 5
iv, _, _ = implied_volatility(
    market_price, S, K, T, r
)
print("New IV:", iv)
```
